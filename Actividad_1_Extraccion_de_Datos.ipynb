```python
#Convertir en dataframe
dataf= pd.read_csv("CountryTable.csv")
dataf.head()
```

```python
#Filtro por objeto
filtro1=dataf[dataf["continent"] == "Asia"]
filtro1
```

```python
#Filtrar por la columna continente y la variable como el continente asiatico.
filtro1.to_csv("Filtro1.csv")
```

```python
#Filtro por filas
filtro2= dataf.iloc[10:20,: ]
filtro2
```

```python
#filtro consecutivo desde la fila 11 hasta la fila 20
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro por columnas
filtro3=dataf.iloc[ : , 2:4]
filtro3
```

```python
#filtro consecutivo que muestra todas las filas pero con los datos desde la tercera columna hasta la cuarta
filtro3.to_csv("Filtro3.csv")
```

```python
dataf= pd.read_csv("CountryTable.csv", index_col=2)#leer el dataframe con un indice
dataf
```

```python
#Filtro de filas vs columnas
filtro4=dataf.loc[["Asia","North America"], ["population"]]
filtro4
```

```python
#filtrar por filas con los datos de "Asia" y "North America" y con la columna de "population"
filtro4.to_csv("Filtro4.csv")
```

```python
#Filtro por cabecera
filtro5=dataf.head(8)
filtro5
```

```python
#mostrar los primeros 8 datos del dataframe
filtro5.to_csv("Filtro5.csv")
```

```python
#Filtro por comparacion
filtro6=dataf[dataf["population"] > 2000000]
filtro6
```

```python
#filtra y muestra los datos que tengan una poblacion mayor a 2000000
filtro6.to_csv("Filtro6.csv")
```

```python
#Filtro Y
filtro7=dataf[(dataf["population"] > 2000000) & (dataf["independence_year"] > 1919)]
filtro7
```

```python
#filtro que muestra los datos con una poblacion mayor a 2000000 y con un a√±o de independencia mayor a 1919
filtro7.to_csv("Filtro7.csv")
```

```python
#Filtro o
filtro8=dataf[(dataf["population"] > 2000000) | (dataf["independence_year"] < 1919)]
filtro8
```

```python
#filtro que muestra los datos con una poblacion mayor a 2000000 o con un a√±o de independencia menor a 1919
filtro8.to_csv("Filtro8.csv")
```

```python
#Filtro not
filtro9=dataf[~(dataf["population"] > 2000000) & ~(dataf["life_expectancy"] < 40 )]
filtro9
```

```python
#filtro que muestra los datos que no tengan una poblacion mayor a 2000000 y que su espectativa de vida no sea menor a 40
filtro9.to_csv("Filtro9.csv")
```
